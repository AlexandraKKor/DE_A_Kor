```python
import time
import psycopg2
from sqlalchemy import create_engine, DATE, Numeric
from datetime import datetime
from sqlalchemy import Date
from sqlalchemy import TIMESTAMP

# Подключение к базе данных
engine = create_engine("postgresql://postgres:postgres@localhost:5432/postgres")
conn = psycopg2.connect("host='localhost' port='5432' dbname='postgres' user='postgres' password='postgres'")
cur = conn.cursor()

# Имя загружаемой таблицы
table_name = 'ft_posting_f'

try:
    # Логирование начала загрузки данных
    cur.execute("INSERT INTO LOGS.ETL_LOGS (start_time, additional_info) VALUES (NOW(), 'Start of loading data into {} table')".format(table_name))
    conn.commit()

    # Пауза на 5 секунд
    time.sleep(5)

    # Очистка таблицы перед загрузкой
    cur.execute("TRUNCATE TABLE ds.ft_posting_f")
    conn.commit()

    # Выгрузка данных из датафрейма ft_posting_f в таблицу ft_posting_f
    ft_posting_f.columns = ft_posting_f.columns.str.lower()
    ft_posting_f.to_sql('ft_posting_f', engine, schema='ds', if_exists='append', index=False, dtype={
    'oper_date': Date,
    'credit_account_rk': Numeric,
    'debet_account_rk': Numeric,
    'credit_amount': Numeric,
    'debet_amount': Numeric
})


    # Логирование успешного окончания загрузки
    cur.execute("INSERT INTO LOGS.ETL_LOGS (end_time, additional_info) VALUES (NOW(), 'End of loading data into {} table')".format(table_name))
    conn.commit()

except Exception as e:
    # Логирование ошибки загрузки данных
    cur.execute("INSERT INTO LOGS.ETL_LOGS (start_time, additional_info) VALUES (NOW(), 'Error loading data into {} table: {}')".format(table_name, str(e)))
    conn.commit()

# Закрываем курсор и соединение
cur.close()
conn.close()
```
